```python
import matplotlib.pyplot as plt
import numpy as np
import pylink
import yaml
```

## 3 dB Antenna

```python
pattern_3dB = pylink.pattern_generator(3)

with open('3dB_pattern.yaml', 'w') as outfile:
    yaml.dump([float(_) for _ in pattern_3dB], outfile)
```

### Plot the result

```python
pattern_3dB_extended = pattern_3dB
pattern_3dB_extended.append(pattern_3dB_extended[0])

angles = [_*(np.pi/180) for _ in range(360)]
angles.append(angles[0])

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6),
                       subplot_kw=dict(polar=True))
 
ax.plot(angles, pattern_3dB_extended, linewidth=1, linestyle='solid')
ax.fill(angles, pattern_3dB_extended, 'skyblue', alpha=0.4)
```
